```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file_1 = "Cleaned_states_all_extended.csv"
Education_Revenue_Totals_df = pd.read_csv(csv_file_1)
Education_Revenue_Totals_df.head()
```

```python
#Clean Years
New_Education_Revenue_Totals_df = Education_Revenue_Totals_df[(Education_Revenue_Totals_df.YEAR >=2000) & (Education_Revenue_Totals_df.YEAR <2017)]
New_Education_Revenue_Totals_df
```

```python
#Remove National and DODEA
Final_New_Education_Revenue_Totals_df = New_Education_Revenue_Totals_df[(New_Education_Revenue_Totals_df.STATE != "NATIONAL") & (New_Education_Revenue_Totals_df.STATE != "DODEA")]
Final_New_Education_Revenue_Totals_df
```

```python
#Rename all column names with all lower case
Final_New_Education_Revenue_Totals_df.columns = Final_New_Education_Revenue_Totals_df.columns.str.lower()
Final_New_Education_Revenue_Totals_df
```

```python
csv_file_2 = "All_Grade_8_Students_copy.csv"
Math_Education_Totals_df = pd.read_csv(csv_file_2)
Math_Education_Totals_df
```

```python
#Reformatting and creating new column for foreign key
Math_Education_Totals_df['Key'] = Math_Education_Totals_df['Year'].apply(str) + '_' +  Math_Education_Totals_df['Jurisdiction'].str.upper()
Math_Education_Totals_df['Year'] = Math_Education_Totals_df['Year'].astype(int)
Math_Education_Totals_df = Math_Education_Totals_df.rename(columns={"Year":"year","Jurisdiction":"jurisdiction","All students": "all_students", "Average scale score":"average_scale_score", "Key":"key"})

Math_Education_Totals_df

```

```python
#Drop National and DODEA
New_Math_Education_Totals_df = Math_Education_Totals_df[(Math_Education_Totals_df.jurisdiction != "National") & (Math_Education_Totals_df.jurisdiction != "DoDEA")]
New_Math_Education_Totals_df
```

```python
connection_string = "admin1:12345@localhost:5432/ETLProject"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
#Connect to PGadmin
Final_New_Education_Revenue_Totals_df.to_sql(name='revenue', con=engine, if_exists='append', index=False)
```

```python
#Connect to PGadmin
New_Math_Education_Totals_df.to_sql(name='math_test_scores', con=engine, if_exists='append', index=False)
```
